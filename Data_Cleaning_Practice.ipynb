```python
##Handling Missing Data
```

```python
df.dropna(subset=["Customer_ID"], inplace=True)
df.isnull().sum()
```

```python
df
```

```python
import re
        
def extract_age(age):
    age_num = re.findall('[0-9]+', str(age))
    if len(age_num)>0:
        return age_num[0]
    else:
        return age 
    
df['Age'].apply(lambda x: extract_age(x))
```

```python
df_age = df[df['Age'] != 'nan years']['Age']
```

```python
# 1. Convert the column to numeric values (handles strings like '52.0' automatically)
df_age_clean = pd.to_numeric(df_age, errors='coerce')

# 2. Calculate the median and convert the final result to an int
age_median = int(df_age_clean.median())
```

```python
age_median
```

```python
df.replace('nan years', age_median, inplace= True)
```

```python
df['Age'].fillna(age_median, inplace=True)
```

```python
df['Age'].unique()
```

```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(), inplace=True)
```

```python
df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in ['Gender','City','Country']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['Last_Purchase_Date'].ffill(inplace=True)
```

```python
import pandas as pd

# 1. Convert strings (like '52.0') to floats, and turn bad data/blanks into NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# 2. Fill any missing values (NaN) with 0 (or a median) so it can safely become an integer
df['Age'] = df['Age'].fillna(0)

# 3. Finally, convert the whole column to integers
df['Age'] = df['Age'].astype('int64')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df_clean[col], color='skyblue', width=0.3)
    plt.title(f'Boxplot of {col}', fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
